## Passo 1
### Listando a tabela para uma melhor vizulização

- A tabela não ficou completa devido ao tamanho mas podemos trabalhar com ela mesmo assim
- Fiz a importação do `pandas` a lib que iremos utilizar hoje
- O `read_excel()` lê a nossa tabela e armazena na variavel `df`
- `display()` mostra nossa tabela na tela

```python
import pandas as pd
df = pd.read_excel(r'vendas.xlsx')
display(df)
```

## Passo 2

### Organizando a tabela e obetendo a quantidade total e o faturamento de cada loja

- Separei a tabela para obter apenas o faturamento total utilizando o conceito de listas do python nativo
- A função `groupyby()` agrupa os dados por uma clona e a função `sum()` realiza a soma das demais colunas
- Ordenei a tabela por quantidade para vemos qual loja vendeu mais utilizando a função `sort_values()`
- `ascending=False` é para ordenar de cima para baixo em ordem decrescente
- A função `reset_index()` foi aplicada para reajustar a tabela após o uso do `groupby()` e obtermos uma vizualização mais clara

```python
df_faturamento = df[['ID Loja', 'Valor Final']].groupby('ID Loja').sum()\
    .sort_values(by='Valor Final', ascending=False).reset_index()

display(df_faturamento)
```

- Mesmo procedimento porém agora para obter a quantidade de vendas

```python
df_quantidade = df[['ID Loja', 'Quantidade']].groupby('ID Loja').sum()\
    .sort_values(by='Quantidade', ascending=False).reset_index()

display(df_quantidade)
```

## Passo 3 
### Calculando o ticket médio

(Precisei pesquisar para entender o que era ticket médio mas consegui pegar o jeito)

- Calculei o ticket médio de cada loja e armazerei uma nova coluna
- A função `to_frame()` indica que estamos criando outra coluna
- Ordenei por ticket médio do maior para o menor


```python
df_ticket = (df_faturamento['Valor Final'] / df_quantidade['Quantidade']).to_frame('Ticket Médio')\
    .sort_values(by='Ticket Médio', ascending=False).reset_index()

display(df_ticket)
```

## Passo bonus
### Tomei a liberdade de inserir o email dos gerentes na tabela

-  Criei emails ficticios mas a inserção na tabela poderia ser feita de várias formas, em um cenário real você poderia inserir o email dos gerentes para facilitar a automação mais a frente
- `apply` indica aplica uma função a cada elemento da serie, neste caso utilizei uma função `lambda` para criar os emails fictios

```python
df_contato = df['ID Loja'].apply(lambda x: f"{x.replace(' ', '').lower()}@empresa.com").to_frame('Contato')
df_contato['ID Loja'] = df['ID Loja']
df_contato = df_contato.groupby(['ID Loja', 'Contato']).sum()

display(df_contato)
```

## Passo 4
### Criando as funções de envio de email
Este passo é o mais chatinho mas vamos lá

- Importei as libs que irei utilizar
- A função `send_email` recebe por parametro um assunto, um recebedor e um conteudo html
- `message.Message()` cria um objeto message para preechimento dos dados
- `slib.SMTP('smtp.gmail.com:587')` configura o serviço smtp
- `starttls()` inicia o tls

- `send_gerencia` recebe como paramentro a tabela, o nome da loja e o email
- Monta o relatorio formalmente e envia o email

- `send_diretoria` recebe como paramentro apenas a tabela de ranking
- Monta o relatorio formalmente e envia o email


este foi um reusmo rapido, você pode obter mais detalhes na documentação do smtplib

```python
import smtplib as slib
from email import message

def send_email(subject, recipient, html_content):
    msg = message.Message()
    msg['From'] = "sciencedata@gmail.com"
    msg['Subject'] = subject
    msg['To'] = recipient
    password = 'sciencedata'
    
    msg.add_header('Content-Type', 'text/html')
    msg.set_payload(html_content)
    
    server = slib.SMTP('smtp.gmail.com:587')
    server.starttls()
    server.login(msg['From'], password)
    server.sendmail(msg['From'], [msg['To']], msg.as_string().encode('utf-8'))
    server.quit()

def send_gerencia(table, name, mail):
    relatorio = f"""
    <p>Prezado, segue o relatório da loja {name}</p>
    {table.to_html()}
    """
    send_email("Relatório de vendas", mail, relatorio)

def send_diretoria(table): 
    relatorio = f"""
    <p>Prezado, segue o ranking de vendas das lojas</p>
    {table.to_html()}
    """
    send_email("Relatório de vendas", "diretoria@email.com", relatorio)
```

## Passo 5
### Enviando de fato os emails

##### Enviando email para diretoria

- Combinei as informações de faturamento, quantidade, ticket médio e contatos em um único DataFrame `df_combined`.
- Ordenei os dados pelo número de vendas para destacar as lojas com melhor desempenho.
- Enviei um email à diretoria com o relatório consolidado.


```python
df_combined = pd.merge(
    pd.merge(
        pd.merge(df_faturamento, df_quantidade, on='ID Loja'),
        df_ticket[['ID Loja', 'Ticket Médio']], on='ID Loja'
    ),
    df_contato, on='ID Loja'
).sort_values(by='Quantidade', ascending=False)

# display(df_combined)

send_diretoria(df_combined)
```

##### Enviando email para gerencia

- Peguei os nomes das lojas
- Pegueis os emails dos gerentes
- Utilizei o loop for para enviar de um a um, o relatório da sua respectiva loja

```python
name = df['ID Loja'].unique()
email = df['Contato'].unique()

for i, s in enumerate(name):
    store = df.loc[df['ID Loja'] == s, ['ID Loja', 'Quantidade', 'Ticket Médio', 'Valor Final']]\
    .groupby('ID Loja').sum()
    
    send_gerencia(store, name[i], email[i])
```
