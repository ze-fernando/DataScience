## Passo 1
### Importando a base de dados e fazendo uns ajustes

- Importei a base de dados csv e utilizei `enconding="ISO-8859-1"` para não ter problemas com acentuação
- A função `drop` e `dropna` são responsaveis por remover algo da tabela, nesse caso o `drop` remove a coluna *CLIENTENUM* e o `dropna` remove as colunas vazias


```python
import pandas as pd

df = pd.read_csv(r'clientes.csv', encoding="ISO-8859-1")

df = df.drop(columns=['CLIENTNUM']).dropna()

display(df)
```

## Passo 2
### Filtrando apenas a categoria de cancelados e analisando a tabela

- Armazenei na variavel `df_cancelados` apenas onde a categoria é cancelado
- A função `info()` tras as informações sobre a tabela em questão

```python
df_cancelados = df[df['Categoria'] == 'Cancelado'].info()

display(df_cancelados)
```

- A função `describe()` faz uma descrição da tabela trazendo media, moda, max, min e etc

```python
df_cancelados = df[df['Categoria'] == 'Cancelado'].describe()

display(df_cancelados)
```

### Atenção especial aqui

- Eu fiz a geração de graficos do proxímo passo e conforme fui analisando, deixei apenas as colunas que fazem sentido pra nossa analise


```python
df_graph = df[['Categoria', 'Categoria Cartão', 'Contatos 12m', 'Qtde Transacoes 12m']]

display(df_graph)
```

## Passo 4
### Criando a função pra gerar os graficos

- Importei a lib que usaremos, neste caso o ploty
- Criei a função para gerar os gráficos já que gerei muitos conforme o topico anterior
- Ela recebe o dataframe e a coluna para a geração, e *Categoria* para `color`
- Geramos o grafico e armazenamos na variável `px`
- A função `px.show()` mostra o gráfico na tela



```python
import plotly.express as px


def new_graph(column, dataframe):
    gp = px.histogram(dataframe, x=column, color='Categoria')
    gp.show()
```

### Vamos para a análise

- Primeiro vamos gerar um gráfico para separar clientes de ex-clientes

```python
new_graph(df_graph['Categoria'], df_graph)
```

- Na categoria do cartão é perceptível que grande parte dos ex-clientes eram da categoria 'blue', o que indica que essa categoria pode não ser tão atraente

```python
new_graph(df_graph['Categoria Cartão'], df_graph)
```

- Na coluna contato percebemos que quanto mais contato o cliente fez mais as chances de cancelar, inclusive todos os clientes com 6 contatos cancelaram o cartão

```python
new_graph(df_graph['Contatos 12m'], df_graph)
```

- Na coluna quantidade de transacoes é notável que quanto menos o cliente usa o cartão, mais as chances dele cancelar, porém a partir de 80 transações o quandro se inverte

```python
new_graph(df_graph['Qtde Transacoes 12m'], df_graph)
```
